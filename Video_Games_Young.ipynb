# Filter

```python
filter_year = 2000

filtered_games = filter(lambda game: game["Year"] > filter_year, data)

filtered_games_list = list(filtered_games)

for game in filtered_games_list:
    print(game)
```

# Map

```python
def convert_sales_to_billions(game):
    game["NA_Sales"] /= 1000
    game["EU_Sales"] /= 1000
    game["JP_Sales"] /= 1000
    game["Other_Sales"] /= 1000
    game["Global_Sales"] /= 1000
    return game

transformed_data = list(map(convert_sales_to_billions, data))

for game in transformed_data:
    print(game)
```
